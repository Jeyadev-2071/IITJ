```python
import pandas as pd
import re
```

```python
with open('data/users.txt', 'r', encoding='utf-8') as file:
            lines = file.readlines()
```

```python
pattern = r'"([^"]+)"'

# Loop through each line
for line in lines:
    line = line.strip()
    parts = re.findall(pattern, line)

    # Create a dictionary to store key-value pairs
    record = {}

    # Iterate through the parts and correctly assign key-value pairs
    i = 0
    while i < len(parts):
        if ':' in parts[i]:
            # If there is a colon, split the key-value and add to the dictionary
            key, value = parts[i].split(':', 1)
            record[key] = value
            i += 1  # Move to the next part
        else:
            # Otherwise, the current part is the key, and the next part is the value
            if i + 1 < len(parts):
                key = parts[i]
                value = parts[i + 1]
                record[key] = value
                i += 2  # Move to the next key-value pair

    # Print the resulting dictionary for each line
    print(record)
```

```python
data = []
```

```python
pattern = r'"([^"]*)"\s+"([^"]*)"'  # Regex pattern to match key-value pairs in quotes

for line in lines:
    line = line.strip()
    if not line:
        continue  # Skip empty lines

    # Find all key-value pairs in the line using regex
    matches = re.findall(pattern, line)
    record = {key: value for key, value in matches}
    data.append(record)

# Convert the list of dictionaries into a DataFrame
user_data = pd.DataFrame(data)
```
